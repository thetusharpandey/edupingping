```python
from fastapi import FastAPI, HTTPException
from fastapi.responses import FileResponse, HTMLResponse

app = FastAPI()

@app.get("/", response_class=HTMLResponse)
async def read_hashgen():
    try:
        return FileResponse("hashGen.html")
    except FileNotFoundError:
        raise HTTPException(status_code=404, detail="File not found")

@app.get("/sadly4041.html", response_class=HTMLResponse)
async def read_sadly4041():
    try:
        return FileResponse("sadly4041.html")
    except FileNotFoundError:
        raise HTTPException(status_code=404, detail="File not found")
        
def check_var_in_hashlog(var: str) -> bool:
    try:
        with open('hashlog.txt', 'r') as file:
            for line in file:
                key, value = line.strip().split(':')
                if key == var:
                    return True
        return False
    except FileNotFoundError:
        return False
    
    
def hash_update(var: str, code: str) -> bool:
    try:
        with open('hashlog.txt', 'r') as file:
            lines = file.readlines()
        flag = 1
        with open('hashlog.txt', 'w') as file:
            for line in lines:
                key, value = line.strip().split(':')
                if value == code:
                    file.write(f"{var}:{value}\n")
                    flag = 0
                else:
                    file.write(line)
            if flag == 1:
                file.write(f"{var}:{code}\n")
                return True
        return True
    except FileNotFoundError: 
        return False

@app.get("/update/{var}/{code}", response_class=HTMLResponse)
async def hash_update_func(var: str, code: str):
    if hash_update(var, code):
        try:
            return FileResponse("success.html")
        except FileNotFoundError:
            raise HTTPException(status_code=500, detail="File not found")
    else:
        return FileResponse("sadly404.html", media_type="text/html")
        raise HTTPException(status_code=404, detail="Variable not found in hashlog")

        
@app.get("/{var}/leads.html", response_class=HTMLResponse)
async def read_hashgen(var: str):
    if check_var_in_hashlog(var):
        try:
            return FileResponse("leads.html")
        except FileNotFoundError:
            raise HTTPException(status_code=404, detail="File not found")
    else:
        return FileResponse("sadly404.html", media_type="text/html")
        raise HTTPException(status_code=404, detail="Variable not found in hashlog")


# Default handler for unmatched paths
@app.get("/{path:path}")
async def catch_all(path: str):
    return FileResponse("sadly404.html", media_type="text/html")

import nest_asyncio
import uvicorn

if __name__ == "__main__":
    nest_asyncio.apply()
    uvicorn.run(app, host="localhost", port=8098)
    
    
    
    
    
""" i do not need to protect this server, As the login links will be created only before the server is 
started. 
hence, when employees come, they sit and get login links on whatsapp/discord, they wait for me to 
start the server and then the work starts. 

This means logins are given, Then work starts. But people will come in between and they will ask for 
the login in that case server can be stopped? yes, or halfday, or started on next breaktime
"""
```
